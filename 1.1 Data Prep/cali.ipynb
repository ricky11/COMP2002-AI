```python
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt

#load the data

california_housing = fetch_california_housing(as_frame=True)

california_housing.data
# california_housing.frame.info()
inputs = california_housing["data"].to_numpy()
targets = california_housing["target"]
# inputs
# inputs


```

```python

california_housing.frame.hist(figsize=(12,10),bins=30,edgecolor="black")
plt.subplots_adjust(hspace=0.7, wspace=0.4)
```

```python


plt.scatter(inputs[:,0], inputs[:,5], c=targets, cmap="viridis")
plt.colorbar()
plt.savefig("california_data.pdf", bbox_inches="tight")
plt.xlabel("MED BLOCK INCOME")
plt.ylabel("avg occ")
plt.show()
```

Normalize

```python
from sklearn.preprocessing import StandardScaler
# Scale the data.
scaler = StandardScaler()
# scaled = scaler.fit_transform(inputs)
longitude = california_housing.data['MedInc']
latitude = california_housing.data['AveOccup']
scaled_x = scaler.fit_transform(longitude.values.reshape(-1, 1))
scaled_y = scaler.fit_transform(latitude.values.reshape(-1, 1))
# print(inputs.ptp(axis=0))
# print(scaled.ptp(axis=0))
plt.scatter(scaled_x, scaled_y, cmap="viridis")
```
